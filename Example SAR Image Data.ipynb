# Code to view SAR datasets
This contains some raw images of the MSTAR, FUSAR-Ship, and OpenSARShip datasets. 

```python
import numpy as np
import matplotlib.pyplot as plt
```

## FUSAR-Ship

```python
data_fusar = np.load('data/Fusar/FusarTrainImages.npz')
print(data_fusar.files)

img_fusar = data_fusar[data_fusar.files[0]]
```

```python
for i in range(3):
    plt.imshow(img_fusar[i, 0, :, :], cmap='gray')
    plt.axis(False)
    plt.savefig('Images/fusar' + str(i) + '.png', transparent=True)
    plt.show()
```

## OpenSARShip

```python
data_opensar = np.load('data/OpenSARShip/SarTrainImages.npz')
print(data_opensar.files)

img_opensar = data_opensar[data_opensar.files[0]]
```

```python
for i in range(3):
    plt.imshow(img_opensar[i, 0, :, :], cmap='gray')
    plt.axis(False)
    plt.savefig('Images/opensar' + str(i) + '.png')
    plt.show()
```
